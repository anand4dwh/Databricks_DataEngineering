```python
df = spark.read.csv("/Volumes/sample_catalog/default/db_catalog/customer.csv", header=True, inferSchema=True)
display(df)
```

```python
df1 = df.filter(df.customer_type == 'Premium')
display(df1)
```

```python
df2 = df.filter(df.country == 'UK')
display(df2) 
```

```python
df_inner = df1.join(df2, on='customer_id', how = 'inner')
```

```python
display(df_inner)
```

```python
df_left = df1.join(df2, on='customer_id', how = 'left')
display(df_left)
```

```python
df_right = df1.join(df2, on='customer_id', how = 'right')
display(df_right)
```

```python
df_outer = df1.join(df2, on='customer_id', how = 'outer')
display(df_outer)
```

```python
from pyspark.sql.functions import broadcast

df3 = df1.join(broadcast(df2), on = 'customer_id', how = 'inner')
```

```python
df3.explain()
```

```python
df1.write.format("parquet").bucketBy(4, "country").saveAsTable("customer_bucket")
```
