```python
# creating a temp view on the customer data csv file

df = spark.read.csv("/Volumes/sample_catalog/default/db_catalog/customer.csv", header=True, inferSchema=True)
display(df)
```

```python
# Creating a Temp view

df.createTempView("customer")
```

```python
df1 = spark.sql("select * from customer")
display(df1)
```

```python
df.createOrReplaceTempView("customer_new")
```

```python
display(spark.sql("show views"))
```

```python
spark.sql("drop view customer_new")
```

```python
display(spark.sql("show views"))
```

```python
# Creating a Global temp views, This view is visible to all users in the workspace
# Since I am using a Serverless compute, I am getting an error, as GLOBAL TEMPORARY VIEW is not supported on serverless compute, but more or less Temp views and Global Temp views are same.
df.createGlobalTempView("customer_global")
```

```python
# This dataframe execution is running on temp view DF

df1 = spark.sql("select customer_id, name, email, customer_type from customer")
display(df1)
            
```

```python
#df2 = spark.sql("select * from customer where customer_type = 'Regular'")
#display(df2)

df3 = spark.sql("select * from customer where customer_type = 'Regular' and country = 'UK'")
display(df3)
```

```python
df4 = spark.sql("select country, count(customer_id) as customer_count from customer group by country")
display(df4)
```

```python
df5 = spark.sql("select country, count(customer_id) as customer_count from customer group by country order by customer_count desc")
display(df5)
```

```python
df6 = spark.sql("select distinct customer_type from customer")
display(df6)
```
